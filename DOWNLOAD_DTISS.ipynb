***04. AMBULATORIAL E HOSPITALAR - DESCOMPACTAR AS MICROBASES EM FORMATO ZIP***

```python
# procedimento de copia da fonte dos arquivos para o diretorio destino

#2023                                                        
n = estado.shape[0]
m = anomes_2023.shape [0]


for k in range(0, 1):       ##estado
    for w in range(0, m):   #anomes
            #########AMBULATORIAL
            # parametros dos diretorios
            OUTPUT_DIR_1 = "C:/DTISS/AMBULATORIAL/"
            OUTPUT_DIR = OUTPUT_DIR_1 + "/" + estado.iloc [k,0] +"/"
            ####AMBULATORIAL consolidado  
            arquivos = estado.iloc [k,0] + '_' + anomes_2023.iloc [w,0] + 'AMB_CONS.zip'
            shutil.unpack_archive(OUTPUT_DIR + arquivos, OUTPUT_DIR)   ##extrair arquivo zip
            ####AMBULATORIAL DETALHADO  
            arquivos = estado.iloc [k,0] + '_' + anomes_2023.iloc [w,0] + 'AMB_DET.zip'
            shutil.unpack_archive(OUTPUT_DIR + arquivos, OUTPUT_DIR)   ##extrair arquivo zip
            #########HOSPITALAR
            # parametros dos diretorios
            OUTPUT_DIR_3 = "C:/DTISS/HOSPITALAR/"
            OUTPUT_DIR_2 = OUTPUT_DIR_3 + "/" + estado.iloc [k,0] +"/"
            ####HOSPITALAR consolidado  
            arquivos = estado.iloc [k,0] + '_' + anomes_2023.iloc [w,0] + 'HOSP_CONS.zip'
            shutil.unpack_archive(OUTPUT_DIR_2 + arquivos, OUTPUT_DIR_2)   ##extrair arquivo zip
            ####HOSPITALAR DETALHADO  
            arquivos = estado.iloc [k,0] + '_' + anomes_2023.iloc [w,0] + 'HOSP_DET.zip'
            shutil.unpack_archive(OUTPUT_DIR_2 + arquivos, OUTPUT_DIR_2)   ##extrair arquivo zip
            print (["---------" + estado.iloc [k,0] +"_"+ anomes_2023.iloc [w,0] + " DESCOMPRIMIDO com sucesso ---------"]) 

```
